```python
# Setting up the Open Ai API key for using the model for text embedding
client = openai.OpenAI(api_key="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")  # Replace with your actual key

#Loading the required article for RAG with an argument to skip unreadable characters
with open("article.txt", "r", encoding="utf-8", errors="ignore") as f:
    full_text = f.read()

# Defining a function to splitt longer chunks into smaller ones
def split_text(text, max_tokens=500):
    
    paragraphs = text.split("\n")
#Defining a variable chunks for storing the final set of chunks    
    chunks = []
#Temporary storage for each chunks    
    current = ""
#Estimating token count    
    token_estimate = lambda x: len(x.split())
#Cleaning up the paragraphs by avoiding empty strings etc
    for p in paragraphs:
        p = p.strip()
        if not p:
            continue
        if token_estimate(current + p) < max_tokens:
            current += " " + p
        else:
            chunks.append(current.strip())
            current = p
    if current:
        chunks.append(current.strip())
    return chunks #returns the list of cleaned chunks
#Returns the number of clenaned chunks created
chunks = split_text(full_text)
print(f"Split into {len(chunks)} chunks.")

# Defines a function to send text to OpenAI and get back a vector
def get_embedding(text):
    response = client.embeddings.create(
        model="text-embedding-ada-002",
        input=[text]
    )
    return response.data[0].embedding

embeddings = [get_embedding(chunk) for chunk in chunks]
print("Embedded all chunks using OpenAI API v1.")

# Setting up a Faiss Index which uses euclidean distance for vector similarity
dimension = len(embeddings[0])
index = faiss.IndexFlatL2(dimension)
index.add(np.array(embeddings).astype("float32"))

# Save the the indices and chunks created
faiss.write_index(index, "act_article_index.faiss")
with open("act_chunks.json", "w", encoding="utf-8") as f:
    json.dump(chunks, f, indent=2)

print("Saved: act_article_index.faiss and act_chunks.json")
```
